Initializing empty dataframe with columns for model analysis

```python
df=pd.DataFrame(columns=["model","prompt","content","completion_tokens","completion_time","prompt_time","queue_time","finish_reason"])
```

```python
# Function to generate text from models
def text_gen_from_models(temp,model_name,prompt):
  llm =ChatGroq(
    temperature=temp,
    groq_api_key="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",
    model_name=model_name
  )
  output=llm.invoke(prompt)
  return output
```

```python
# Function to retrieve information from models and return a DataFrame row
def get_from_models(model,prompt,temp):
  final_output=text_gen_from_models(temp=temp,model_name=model,prompt=prompt)
  token=final_output.response_metadata['token_usage']
  new_row = {
        "model": model,
        "temperature": temp,
        "prompt": prompt,
        "content": final_output.content,
        "completion_tokens": token["completion_tokens"],
        "completion_time": token["completion_time"],
        "prompt_time": token["prompt_time"],
        "queue_time": token["queue_time"],
        "finish_reason": final_output.response_metadata["finish_reason"]
    }
  return new_row
```

```python
# List of models and prompts for processing
model_list=["llama-3.3-70b-versatile","gemma2-9b-it","llama-3.1-8b-instant","mixtral-8x7b-32768","qwen-2.5-32b"]
prompt_list=["Explain the concept of deep learning and its applications in modern technology.",
             "what is love?",
             "how can I start earning as 15 year old?",
             "How can AI be used to tackle global challenges like climate change?"
             ]
# List to hold the rows before appending to DataFrame
rows=[]
# for Looping through models and prompts to collect results
for model in model_list:
    for prompt in prompt_list:
        if not prompt:  # just to skip the empty prompts
            continue

        # Getting results for both temperature values
        new_row_0 = get_from_models(model, prompt, temp=0)
        new_row_1 = get_from_models(model, prompt, temp=1)

        # Appending results to the rows list
        rows.append(new_row_0)
        rows.append(new_row_1)
```

```python
#adding the rows to the final actual dataframe
df = pd.DataFrame(rows)
```
